To get us familiar with Python coding let's work on an example problem of how probability can be applied to biology and biological analysis.  

The use of DNA and sequencing has changed dramatically over the decades.  In the 1980s when DNA was first used in Forensics, we couldn't yet do extensive sequencing and instead the DNA was "typed" for a few key spots in the genome known as "markers".  The type of markers that were collected is known as " variable number tandem repeats" VNTRs.  These are sections of DNA that are made up of repeated sequences of **bases** ("ACGTs")  where each repeat is about 10-100 bases long.  The number of times that the focal sequences repeated varied among individuals.  For example, some individuals may carry 10 "repeats" and others "14".  

Suppose that at a given marker there are 4 different **alleles** (e.g., alleles refer to the different genetic "options" that an individual can have). Where the number of repeats in each allele and the frequency of that allele in the population is given by:

- Allele 1: 10 repeats, $p_1=0.1$
- Allele 2: 12 repeats, $p_2=0.3$
- Allele 3: 15 repeats, $p_3=0.35$
- Allele 4: 17 repeats, $p_4=0.25$

**1.  Use a histogram to show the relative probabilities of each allele in the population.**

```python
import matplotlib.pyplot as plt

# Probabilities
probabilities = [0.1, 0.3, 0.25, 0.35]

# Labels for the bars
categories = ['Allele 1', 'Allele 2', 'Allele 3', 'Allele 4']

# Plotting the bar chart
plt.bar(categories, probabilities, color=['blue', 'green', 'orange', 'red'])

# Adding labels and title
plt.xlabel('Alleles')
plt.ylabel('Probabilities')
# plt.title('Bar Chart of Probabilities')

# Display the plot
plt.show()
```
